```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
import sklearn
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix,classification_report 
```

```python
#read and scale data
df1 = pd.read_csv("/home/just/Scaricati/clustering.csv")
names1 = df1.columns
scaler = preprocessing.StandardScaler()
scaled_df1 = scaler.fit_transform(df1)
scaled_df1 = pd.DataFrame(scaled_df1, columns=names1)
X = np.array(scaled_df1)
plt.scatter(X[:,0],X[:,1])
plt.xlabel('timestamp')
plt.ylabel('power')
plt.title('power vs timestamp with outliers')
plt.show()


```

```python
#build and run model
clustering = KMeans(n_clusters = 3, random_state = 0)
clustering.fit(X)
```

```python
print(clustering.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=clustering.labels_,  cmap='rainbow')
plt.xlabel('timestamp')
plt.ylabel('power')
plt.title('power vs timestamp with outliers')
plt.show()
```
